```python
import pandas as pd
import numpy as np
```

```python
#Carga del csv a manipular
accidentes_bicicletas=pd.read_csv("./AccidentesBicicletas_2021.csv", sep=";")
```

```python
accidentes_bicicletas
```

```python
#Agrupar por el tipo de accidente
accidentes_por_tipo = accidentes_bicicletas.groupby(by="tipo_accidente")
for accidente in accidentes_por_tipo:
    print(accidente)
```

```python
#Saca el máximo y mínimo de lesividad
accidentes_bicicletas.agg({'cod_lesividad': ['max','min']})
```

```python
#Agrupar accidentes por el rango de edad
accidentes_por_edad = accidentes_bicicletas.groupby(by="rango_edad").size()
print(accidentes_por_edad)
```
